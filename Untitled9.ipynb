```python
import pandas as pd
import numpy as np

np.random.seed(42)

data = {
    'company': [
        'LIC', 'Delhivery', 'Zomato', 'Nykaa', 'Paytm',
        'Go Digit', 'Ola Electric', 'Bajaj Housing', 'NTPC Green', 'Hyundai India',
        'Mamaearth', 'Innova Captab', 'Tata Tech', 'IREDA', 'Cello World',
        'JSW Infrastructure', 'Mankind Pharma', 'Nexus Select', 'Zaggle', 'Sai Silks'
    ],
    'sector': [
        'Insurance', 'Logistics', 'Food Tech', 'E-commerce', 'Fintech',
        'Insurance', 'EV', 'Finance', 'Energy', 'Auto',
        'FMCG', 'Pharma', 'Auto', 'Energy', 'Consumer',
        'Infrastructure', 'Pharma', 'Real Estate', 'Fintech', 'Retail'
    ],
    'issue_price': [
        949, 487, 76, 1125, 2150,
        272, 76, 70, 108, 1960,
        324, 448, 500, 32, 648,
        113, 1080, 100, 164, 222
    ],
    'listing_price': [
        867, 493, 116, 2001, 1564,
        321, 91, 150, 111, 1934,
        330, 502, 1200, 75, 802,
        143, 1300, 109, 204, 181
    ],
    'subscription_times': [
        2.9, 1.7, 38.2, 81.8, 1.9,
        11.9, 4.3, 63.6, 2.3, 2.4,
        3.2, 35.2, 69.4, 38.8, 44.8,
        37.4, 41.6, 6.1, 22.3, 3.5
    ],
    'issue_size_cr': [
        21008, 5235, 9375, 5352, 18300,
        2615, 6146, 6560, 10000, 27870,
        1701, 570, 3042, 2150, 1900,
        2800, 4326, 3200, 563, 1201
    ],
    'year': [
        2022, 2022, 2021, 2021, 2021,
        2024, 2024, 2024, 2024, 2024,
        2023, 2023, 2023, 2023, 2023,
        2023, 2023, 2023, 2023, 2022
    ]
}

df = pd.DataFrame(data)

# Listing gain % calculate பண்ணு
df['listing_gain_%'] = ((df['listing_price'] - df['issue_price']) / df['issue_price'] * 100).round(2)

# Winner / Loser label
df['status'] = df['listing_gain_%'].apply(lambda x: 'Winner' if x > 0 else 'Loser')

print("✅ Dataset ready!")
print(df[['company', 'issue_price', 'listing_price', 'listing_gain_%', 'status']].to_string())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('india_ipo.csv.csv')  # உன் file name இதுதான் — சரியா இருக்கு
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst look:")
df.head()
```

```python
# Clean column names
df.columns = df.columns.str.strip()

# Date convert
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')
df['Year'] = df['Date'].dt.year

# Numeric columns clean
df['Listing Gains(%)'] = pd.to_numeric(df['Listing Gains(%)'], errors='coerce')
df['Current  Gains (%)'] = pd.to_numeric(df['Current  Gains (%)'], errors='coerce')
df['Issue Size (in crores)'] = pd.to_numeric(df['Issue Size (in crores)'], errors='coerce')
df['Total'] = pd.to_numeric(df['Total'], errors='coerce')

# Winner / Loser label
df['Status'] = df['Listing Gains(%)'].apply(lambda x: 'Winner' if x > 0 else 'Loser')

print("✅ Clean complete!")
print(f"Total IPOs: {len(df)}")
print(f"Winners: {(df['Status']=='Winner').sum()}")
print(f"Losers: {(df['Status']=='Loser').sum()}")
print(f"\nYears covered: {sorted(df['Year'].dropna().unique().astype(int).tolist())}")
print(f"\nAvg Listing Gain: {df['Listing Gains(%)'].mean():.2f}%")
print(f"Best IPO: {df.loc[df['Listing Gains(%)'].idxmax(), 'IPO Name']} — {df['Listing Gains(%)'].max():.2f}%")
print(f"Worst IPO: {df.loc[df['Listing Gains(%)'].idxmin(), 'IPO Name']} — {df['Listing Gains(%)'].min():.2f}%")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('India IPO Analysis 2021–2024', fontsize=18, fontweight='bold', y=1.01)

# Chart 1 — Top 10 IPOs by Listing Gain
top10 = df.nlargest(10, 'Listing Gains(%)')
colors = ['#2ecc71' if x > 0 else '#e74c3c' for x in top10['Listing Gains(%)']]
axes[0,0].barh(top10['IPO Name'], top10['Listing Gains(%)'], color=colors)
axes[0,0].set_title('Top 10 IPOs — Listing Gain %', fontweight='bold')
axes[0,0].set_xlabel('Listing Gain %')
axes[0,0].axvline(0, color='black', linewidth=0.8)

# Chart 2 — Winner vs Loser Pie
counts = df['Status'].value_counts()
axes[0,1].pie(counts, labels=counts.index, autopct='%1.1f%%',
              colors=['#2ecc71','#e74c3c'], startangle=90,
              wedgeprops={'edgecolor':'white', 'linewidth':2})
axes[0,1].set_title('Winner vs Loser IPOs', fontweight='bold')

# Chart 3 — Subscription vs Listing Gain Scatter
axes[1,0].scatter(df['Total'], df['Listing Gains(%)'],
                  color='#3498db', alpha=0.6, s=60)
axes[1,0].set_title('Subscription Rate vs Listing Gain', fontweight='bold')
axes[1,0].set_xlabel('Subscription (times)')
axes[1,0].set_ylabel('Listing Gain %')
axes[1,0].axhline(0, color='red', linewidth=0.8, linestyle='--')

# Chart 4 — Year wise Avg Listing Gain
yearly = df.groupby('Year')['Listing Gains(%)'].mean().round(2)
bar_colors = ['#3498db','#9b59b6','#e67e22','#1abc9c']
axes[1,1].bar(yearly.index.astype(int), yearly.values, color=bar_colors[:len(yearly)])
axes[1,1].set_title('Avg Listing Gain by Year', fontweight='bold')
axes[1,1].set_xlabel('Year')
axes[1,1].set_ylabel('Avg Gain %')
axes[1,1].axhline(0, color='red', linewidth=0.8, linestyle='--')
for i, (yr, val) in enumerate(yearly.items()):
    axes[1,1].text(yr, val + 0.5, f'{val:.1f}%', ha='center', fontsize=10)

plt.tight_layout()
plt.savefig('ipo_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Charts saved as ipo_analysis.png")
```

```python
print("🏆 TOP 5 IPOs (Listing Gain):")
print(df.nlargest(5, 'Listing Gains(%)')[['IPO Name','Issue','Listing Gains(%)','Year']].to_string(index=False))

print("\n❌ WORST 5 IPOs:")
print(df.nsmallest(5, 'Listing Gains(%)')[['IPO Name','Issue','Listing Gains(%)','Year']].to_string(index=False))

print("\n📅 Year-wise Summary:")
print(df.groupby('Year').agg(
    Total_IPOs=('IPO Name','count'),
    Avg_Listing_Gain=('Listing Gains(%)','mean'),
    Winners=('Status', lambda x: (x=='Winner').sum())
).round(2))
```
